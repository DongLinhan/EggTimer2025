```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def process_image(image_path, threshold_value, save_threshold_path, save_highlighted_path):
    # Read the image in grayscale
    image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
    if image is None:
        print("Error: Unable to read image.")
        return

    # Apply thresholding
    _, thresholded = cv2.threshold(image, threshold_value, 255, cv2.THRESH_BINARY_INV)

    # Count the number of dark pixels
    dark_pixel_count = np.sum(thresholded == 255)

    # Visualize the results
    print(f"Number of dark pixels (below threshold {threshold_value}): {dark_pixel_count}")
    
    # Save the thresholded image
    cv2.imwrite(save_threshold_path, thresholded)
    print(f"Thresholded image saved at: {save_threshold_path}")
    
    # Create a mask for visualization (highlighted in red)
    dark_pixels_colored = cv2.merge([thresholded, thresholded, thresholded])  # Convert grayscale to 3-channel
    dark_pixels_colored[:, :, 1:] = 0  # Highlight dark pixels in red for visualization
    
    # Save the highlighted image
    cv2.imwrite(save_highlighted_path, dark_pixels_colored)
    print(f"Highlighted dark pixels image saved at: {save_highlighted_path}")

    # Display the original image and thresholded result
    plt.figure(figsize=(12, 6))
    plt.subplot(1, 3, 1)
    plt.title("Original Image")
    plt.imshow(image, cmap='gray')
    plt.axis('off')

    plt.subplot(1, 3, 2)
    plt.title("Thresholded (Dark Pixels)")
    plt.imshow(thresholded, cmap='gray')
    plt.axis('off')

    plt.subplot(1, 3, 3)
    plt.title("Highlighted Dark Pixels")
    plt.imshow(cv2.cvtColor(dark_pixels_colored, cv2.COLOR_BGR2RGB))
    plt.axis('off')

    plt.tight_layout()
    plt.show()

# Specify paths and threshold value
image_path = "/Users/donglinhan/Desktop/EHyper2025//EggCounter_images/NCS/LVP-L3.jpg"  # Replace with your image file path
threshold_value = 170  # Adjust threshold based on your needs
save_threshold_path = "thresholded_image_control.png"  # Path to save the thresholded image
save_highlighted_path = "highlighted_dark_pixels_ctl.png"  # Path to save the highlighted image

# Process the image
process_image(image_path, threshold_value, save_threshold_path, save_highlighted_path)
```
